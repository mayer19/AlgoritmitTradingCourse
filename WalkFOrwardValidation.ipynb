```python
#Imports
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
df = pd.read_csv("DataNExcels/df_microdt_linkedin.csv").set_index('Date')
df = df.drop(['Adj Close', 'change_direction'], axis=1)
df
```

```python
#How to use Walk Forward Validation
from sklearn.model_selection import TimeSeriesSplit

ts = TimeSeriesSplit(test_size=200)
```

```python
splits = ts.split(X=df)
splits
```

```python
#create list to append our split data
list_df_train = []
list_df_test = []

for index_train, index_test in ts.split(df):
    list_df_train.append(df.iloc[index_train])
    #print('Train', df.iloc[index_train])
    list_df_test.append(df.iloc[index_test])
    #print('Test', df.iloc[index_test])
```

```python
#Create x and Y
y = df['percentual_change']
X = df[['Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
#create list to append our split data
list_df_train = []
list_df_test = []

for index_train, index_test in ts.split(df):
    X_train, y_train = X.iloc[index_train], y.iloc[index_train]
    X_test, y_test = X.iloc[index_test], y.iloc[index_test]
```

```python
model_dt = DecisionTreeRegressor(max_depth=15, random_state=42)
model_dt.fit(X_train, y_train)

y_pred = model_dt.predict(X_test)
error_mse = mean_squared_error(y_test, y_pred) #compare predictions with test data values
error_mse
```

```python
#Loop to train in all the data based in our 5 splits using walk forvard validation
model_dt = DecisionTreeRegressor(max_depth=15, random_state=42)

error_mse_list = []

for index_train, index_test in ts.split(df):
    X_train, y_train = X.iloc[index_train], y.iloc[index_train]
    X_test, y_test = X.iloc[index_test], y.iloc[index_test]

    model_dt.fit(X_train, y_train)
    
    y_pred = model_dt.predict(X_test)
    error_mse = mean_squared_error(y_test, y_pred)
    
    error_mse_list.append(error_mse)
```
